```python
import pandas as pd


try:
    # Ler o arquivo Parquet em um DataFrame
    df = pd.read_parquet("OriginaisNetflix.parquet")

    # Transformar a coluna Premiere de string em datetime
    df["Premiere"] = pd.to_datetime(df["Premiere"])

    # Transformar coluna dt_inclusao de string em datetime
    df['dt_inclusao'] = pd.to_datetime(df['dt_inclusao'])

    # Exibir apenas a data (ano-mês-dia)
    df['dt_inclusao'] = df['dt_inclusao'].dt.date

    # Criar uma coluna temporária para ordenação
    df['Sort'] = df['Active'].astype(str) + df['Genre']

    # Deixar toda inicial maiuscula do campo genre 
    df['Genre'] = df['Genre'].str.capitalize()

    # Ordenar os dados com base na coluna temporária e no campo Genre
    df = df.sort_values(['Sort', 'Genre'], ascending=[False, True], key=lambda x: x.astype(str).str.casefold())

    # Remover a coluna temporária
    df = df.drop('Sort', axis=1)

    # Removendo dados duplicados do dataframe
    df = df.drop_duplicates()

    # Alterando as linhas de Tba para a ser anunciado
    df['Seasons'] = df['Seasons'].replace('TBA', 'a ser anunciado')

    # Criação da coluna Timestamp
    df['Data da Alteração'] = pd.Timestamp.now()

    # Dicionário mapeamento dos nomes das colunas
    mapeamento = {'Title': 'Título', 'Genre': 'Gênero', 'GenreLabels': 'EtiquetasDeGênero',
                  'Premiere': 'Estreia', 'Seasons': 'Temporadas', 'SeasonsParsed': 'Temporadas Interpretadas',
                  'EpisodesParsed': 'EpisódiosInterpretados', 'Length': 'Duração', 'MinLength': 'DuraçãoMínima',
                  'MaxLength': 'DuraçãoMáxima', 'Status': 'Status', 'Active': 'Ativo', 'Table': 'Tabela',
                  'Language': 'Idioma'}

    # Alterar os nomes das colunas
    df.columns = [mapeamento.get(coluna, coluna) for coluna in df.columns]

    print("Execução concluída sem erros.")

    # Selecionar apenas as colunas desejadas
    colunas_desejadas = ['Título', 'Gênero', 'Temporadas', 'Estreia', 'Idioma', 'Ativo', 'Status', 'dt_inclusao', 'Data da Alteração']
    df_csv = df[colunas_desejadas]

    # Salvar o DataFrame em um arquivo CSV
    df_csv.to_csv('OriginaisNetflix.csv', sep=';', index=False)

    display(df_csv)

except Exception as e:
    print("Ocorreu um erro durante a execução:", str(e))
```
